```python
%sql
--Movies Table with Cleaned IMDb Ratings

CREATE OR REPLACE TABLE workspace.default.movies_clean AS
SELECT 
    title,
    industry,
    TRY_CAST(imdb_rating AS DOUBLE) AS imdb_rating
FROM workspace.default.movies;
```

```python
%sql

--Structure of movies_clean Table

DESCRIBE workspace.default.movies_clean;
```

```python
%sql
-- top 5 movies with highest imdb_rating (excluding NULL values)

SELECT title, imdb_rating
FROM workspace.default.movies_clean
WHERE imdb_rating IS NOT NULL
ORDER BY imdb_rating DESC
LIMIT 5;
```

```python
%sql
-- average IMDb rating per industry

select industry,avg(imdb_rating)as avg_rating from workspace.default.movies_clean
group by industry;
```

```python
%sql

--total revenue generated by each industry
select industry,sum(revenue) as total_revenue from workspace.default.movies_clean
group by industry;
```

```python
%sql

--movies per language
select language,count(*) from workspace.default.movies_clean
group by language;
```

```python
%sql

--movies with rating greater than 8
select * from workspace.default.movies_clean
where imdb_rating>8;
```

```python
%sql
--top 5 most profitable movies
select * from workspace.default.movies_clean
order by profit desc
limit 5
```

```python
%sql
--top 5 most profitable movies
select * from workspace.default.movies_clean
order by profit desc;
```

```python
%sql

--movies with loss (profit < 0)

select * from workspace.default.movies_clean
where profit<=0;
```

```python
%sql

--average profit per industry

select industry,avg(profit)as avg_profit from workspace.default.movies_clean
group by industry;
```

```python
%sql
select distinct currency from workspace.default.movies_clean;
```

```python
%sql

--total profit by each currency
select currency,sum(profit)as total_profit from workspace.default.movies_clean
group by currency;
```

```python
%sql

select * from workspace.default.movies_clean
where budget is null or revenue is null;
```

```python
%sql

--rating ≥ 8 → Hit
-- 5–8 → Average
-- <5 → Flop

select *,case 
    when imdb_rating>=8 then 'HIT'
    when imdb_rating>=5 then 'AVERAGE'
    else 'FLOP'
    end as rating
from workspace.default.movies_clean;
```

```python
%sql

select case 
    when imdb_rating>=8 then 'HIT'
    when imdb_rating>=5 then 'AVERAGE'
    else 'FLOP'
    end as rating,count(*) as movies_count
from workspace.default.movies_clean
group by rating
```

```python
%sql

select *, dense_rank() over (partition by industry order by imdb_rating desc) as ranked from workspace.default.movies_clean
```

```python
%sql

--year-wise average rating

select release_year,avg(imdb_rating)as avg_rating from workspace.default.movies_clean
group by release_year
order by release_year;
    
```

```python
%sql

--industry with highest total revenue

select industry,sum(revenue)as total_revenue from workspace.default.movies_clean
group by industry
order by total_revenue desc;
```

```python
%sql

--Rank movies based on revenue
SELECT
  title,
  revenue,
  DENSE_RANK() OVER (ORDER BY revenue DESC) AS revenue_rank
FROM
  workspace.default.movies;
```

```python
%sql
--ROI :the movie earned how much profit of what was invested

select *,(revenue/budget)*100 as ROI from workspace.default.movies_clean;

--ex:
-- ROI=200/300×100=150%

--  That means the movie earned 1.5x profit of what was invested
```

```python
%sql

select *,(revenue/budget)*100 as ROI from workspace.default.movies_clean
order by ROI desc
limit 1;
```

```python
%sql

--top studio based on total revenue

select studio,sum(revenue) as total_revenue from workspace.default.movies_clean
group by studio
order by total_revenue desc
```
